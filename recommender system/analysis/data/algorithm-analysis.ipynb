```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def factor(data):
    '''
    :param data: Dataframe input 
    :return: three dataframe from each level_1 parameter
    '''
    test_time = data[data['level_1']=='test_time']
    fit_time = data[data['level_1']=='fit_time']
    test_rmse = data[data['level_1']=='test_rmse']
    print('fit_time,test_rmse,test_time')
    return fit_time,test_rmse,test_time
```

```python
surprise_100k = pd.read_csv("./data/100k-data.csv",index_col=False)
surprise_100k = surprise_100k.iloc[:,1:]
surprise_1m = pd.read_csv("./data/1m-data.csv",index_col=False)
surprise_1m = surprise_1m.iloc[:,1:]
```

```python
a,b,c = factor(surprise_100k)
```

```python
sns.lineplot(data=a.reset_index().iloc[:,1:])
plt.title('fitting time')
```
